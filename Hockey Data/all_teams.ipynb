```python
import pandas as pd
import numpy as np
```

```python
# Read in CSV
all_teams = pd.read_csv('all_teams.csv')
all_teams
```

```python
# Filter by home team only to remove duplicates

all_teams = all_teams[all_teams['home_or_away'] == 'HOME']
all_teams
```

```python
# Filter by 'all' situation to get game statistics by all situations

all_teams = all_teams[all_teams['situation'] == 'all']
all_teams
```

```python
# Create win column

# Eliminate Shootout Games

all_teams = all_teams[(all_teams['goalsFor'] - all_teams['goalsAgainst'] != 0)]

all_teams['isWin'] = (all_teams['goalsFor'] - all_teams['goalsAgainst'] > 0) + 0

all_teams
```

```python
# Regular season only
all_teams = all_teams[all_teams['playoffGame'] == 0]
all_teams
```

```python
# Remove corrupted/empty games
all_teams.dropna(inplace=True)
all_teams
```

```python
# Drop unneccessary variables (keeping season for test/train split purposes)
all_teams = all_teams.drop(['situation', 'name', 'playerTeam', 'home_or_away', 'position', 'gameDate', 'playoffGame', 'gameId'], axis=1)
all_teams
```

```python
all_teams.to_csv('all_teams_cleaned.csv', index=False)
```
